```python
import numpy as np
import pandas as pd
import os
```

```python
print('### Marketing Funnel by Olist ###')
for idx, file in enumerate(os.listdir('Data/Marketing')):
    print(idx, '-', file)
print('\n---------------------------------------------\n')

print('### Brazilian E-Commerce Public Dataset by Olist ###')
for idx, file in enumerate(os.listdir('Data/E-Commerce')):
    print(idx, '-', file)
```

## Marketing data set

```python
mql = pd.read_csv('Data/Marketing/olist_marketing_qualified_leads_dataset.csv')
cd = pd.read_csv('Data/Marketing/olist_closed_deals_dataset.csv')
```

```python
# marketing funnel dataset (NaNs are leads that did not close a deal)
mf = mql.merge(cd, on='mql_id', how='left')
mf.head(10)
```

## E-commerce Dataset

```python
sellers = pd.read_csv('Data/E-Commerce/olist_sellers_dataset.csv')
```

```python
# marketing funnel merged with sellers (this way you get seller location)
mf_sellers = mf.merge(sellers, on='seller_id', how='left')
```

```python
# order items dataset
items = pd.read_csv('Data/E-Commerce/olist_order_items_dataset.csv')
```

```python
# marketing funnel merged with items (this way you get products sold by sellers)
products_by_sellers = mf.merge(items, on='seller_id', how='left')
```

```python
products_by_sellers['Closed'] = products_by_sellers['seller_id'].isna().map(lambda x : 0 if x==True else 1)
```

```python
products_by_sellers.to_csv('products_by_sellers.csv', index=False)
```
